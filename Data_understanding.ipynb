## Removing cousines with < 50 recipes

```python
recipes_counts = recipes["cuisine"].value_counts()
cuisines_indices = recipes_counts > 50
cuisines_indices
```

```python
# give all cuisine name as number array
cuisine_name_array =recipes_counts.index.values
cuisine_name_array
```

```python
# conver the true/faulse series to number array
cuisines_indices_array = np.array(cuisines_indices)
cuisines_indices_array
```

```python
# Use the boolean mask to select only cuisine with true

cuisine_name_greater_50 = np.array(cuisine_name_array)[cuisines_indices_array]
cuisine_to_keep = list(cuisine_name_greater_50)
cuisine_to_keep
```

```python
rows_before = recipes.shape[0]
print("Number of rows of original dataframe is {}.".format(rows_before))
recipes = recipes.loc[recipes['cuisine'].isin(cuisine_to_keep)]
rows_after = recipes.shape[0]

print("Number of rows of processed dataframe is {}.".format(rows_after))

print("{} rows removed!".format(rows_before - rows_after))
```

###  The data is bias towards America ingridetnt because there are about 40000 America recipe

Creating profire for each cuisine

```python
cuisines = recipes.groupby("cuisine").mean()
cuisines.head()
```

```python
num_ingredient = 4   # define number of top ingredient to print

# define a function that prints the top ingredient for each cuisine

def print_top_ingredients(row):
    print(row.name.upper())
    row_sorted = row.sort_values(ascending=False)*100
    top_ingredients = list(row_sorted.index.values[0:num_ingredient])

    for ind, ingridient in enumerate(top_ingredients):
        print("%s(%d%%) " % (ingridient, row_sorted[ind]), end='')
    print("\n" )

create_cuisines_profile = cuisines.apply(print_top_ingredients, axis=1)
```
